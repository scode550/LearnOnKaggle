```python
df=pd.read_csv("/kaggle/input/fake-news-classification/WELFake_Dataset.csv")
df=df.drop(["Unnamed: 0"], axis = 1)

df.dropna(inplace=True)
df.reset_index(drop=True,inplace=True)

reg_df=df.replace(to_replace='[\\\\,./?<>{}\\[\\]!@#$%^&*\\-+=;:â€™\\"\'~`]',value='',regex=True)
reg_df.head()
```

```python
print(reg_df.shape)
```

```python
X=reg_df['text']
y=reg_df['label']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print(X_train.shape,X_test.shape)
vectorized=TfidfVectorizer()
analyze = vectorized.build_analyzer()
# print(analyze(X_train[1234]))
vectorized_X_train=vectorized.fit_transform(X_train)
vectorized_X_test=vectorized.transform(X_test)
```

```python
print(vectorized_X_train.shape,vectorized_X_test.shape)
```

```python
models=[LogisticRegression(max_iter=1000),KNeighborsClassifier(max_iter=1000),DecisionTreeClassifier(max_iter=1000),SVC(max_iter=1000),GaussianNB()]
```

```python
for currentModel in models:
    scores=cross_val_score(currentModel,vectorized_X_train,y_train,cv=3)
    print(f"Cross-Validation Score of {currentModel}: ",scores.mean())
```
